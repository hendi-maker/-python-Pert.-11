## Pembacaan berkas
Alur pembacaan berkas :
1. Buka berkas
2. Operasi berkas (baca/tulis)
3. Tutup berkas

Cara pembukaan berkas :
    
    file_object = open(namafile,action)

|Jenis Action|Arti                      |
|------------|--------------------------|
|r           |Baca berkas yang sudah ada|
|w           |Simpan ke berkas. Jika berkas tidak ada, berkas akan dibuat|
|a           |Tambahkan data ke akhir berkas|
|r+           |Baca dan tulis ke berkas yang sudah ada|
|w+           |Seperti 'a' namun akan membuat berkas terlebih dahulu jika tidak ada|

```python
#baca berkas
berkas = open('berkas1.txt','r')
isi_berkas = berkas.read() # baca data sampai akhir
berkas.close()

print(isi_berkas)
```

```python
#baca berkas pada beberapa data dalam jumlah tertentu
berkas = open('berkas1.txt','r')
print(berkas.read(4)) # baca 4 data pertama
print(berkas.read(6)) # baca 6 data berikutnya (termasuk spasi)
```

```python
#Tulis Berkas
with open("test.txt", "w") as f:
    f.write("The new first line\n")
    f.write("The new second line\n")
f.close()

#baca berkas
berkas = open('test.txt','r')
isi_berkas = berkas.read() # baca data sampai akhir
berkas.close()
print(isi_berkas)
```

```python
#Tulis Berkas dgn append
with open("test.txt", "a") as f:
    f.write("The new third line\n")
f.close()

#baca berkas
berkas = open('test.txt','r')
isi_berkas = berkas.read() # baca data sampai akhir
berkas.close()
print(isi_berkas)
```

```python
#Tulis berkas
berkas1 = open('berkas1.txt','r')
berkas2 = open('berkas2.txt','r')

isi_berkas1 = berkas1.read()
isi_berkas2 = berkas2.read()

berkas1.close()
berkas2.close()

isi_gabungan = isi_berkas1 + "\n" + isi_berkas2

berkas_tulis = open('berkas3.txt','w')
berkas_tulis.write(isi_gabungan)
berkas_tulis.close()

# menggunakan try..finally
try:
    berkas3 = open('berkas3.txt','r')
    # melakukan beberapa operasi
    isi_berkas3 = berkas3.read()
    isi_gabungan2 = isi_berkas3 + "\n" + isi_berkas3
    #isi_gabungan2 = isi_berkas3 + "\n" + isi_berkas5
    berkas_tulis = open('berkas4.txt','w')
    berkas_tulis.write(isi_gabungan2)
finally:
    berkas_tulis.close()
```

### Membaca dan menulis dalam mode biner
Mode biner kita gunakan saat berhubungan dengan file gambar atau file .exe atau yg lainnya

```python
import binascii

f = open("img/thumbs-up.png", 'r+b')
img_thumbs = f.read()
f.close()
#print (binascii.hexlify(ima))

f = open("file-thumbs-up.txt", "w") #simpan dalam file, misal "file-thumbs-up.txt"
f.write(binascii.hexlify(img_thumbs).decode('utf-8'))
f.close()

f = open("file-thumbs-up.txt", "r")
getbyte = f.read()
f.close()
#print(getbyte)

# Save it as a binary image, write getbytes:
getbytes = (binascii.unhexlify(getbyte))
f = open("thumbs-up-write.png", "w+b")
f.write(getbytes)
f.close

# show image
import matplotlib.pyplot as plt
plt.box(False) # remove box

# remove all the ticks and label each bar
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='off')
from scipy.misc import imread
plt.imshow(imread("img/thumbs-up.png"))
plt.show()
plt.close()
```

```python
# show image subplot
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

img1 = mpimg.imread('img/thumbs-up.png')
w=30
h=30
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 4))

columns = 3
rows = 1
for i in range(1, columns*rows +1):
    if i==1:        
        img = img1
    elif i==2:
        img = img1[:,:,0]
    else:
        img = Image.open('img/thumbs-up.png').convert('LA')
        #img.thumbnail((64, 64), Image.ANTIALIAS) # resizes image in-place
    plt.subplot(rows, columns, i)     
    plt.axis('off')
    plt.imshow(img) 
    
plt.show()
plt.close()
```
